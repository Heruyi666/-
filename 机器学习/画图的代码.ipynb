```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import matplotlib.font_manager as font_manager

Clean = [0.324,0.382,0.300]
DS = [0.274,0.225,0.236]
PDSR =  [0.221,0.180,0.176]

labels = ['ML-1M','ML-20M','LastFM']
fontsize=25

x = np.arange(len(labels))
width = 0.8
plt.figure(figsize=(10, 8))
rects1 = plt.bar(x - width/4 , Clean, width=width/4, color='#FF9848')
rects2 = plt.bar(x  , DS, width=width/4, color='#1B2ACC')
rects3 = plt.bar(x + width/4, PDSR, width=width/4, color='g')

plt.ylim(0, 0.5)

plt.bar_label(rects1, label_type='edge', rotation=85, fontsize=20)
plt.bar_label(rects2, label_type='edge', rotation=85, fontsize=20)
plt.bar_label(rects3, label_type='edge', rotation=85, fontsize=20)

plt.yticks(fontsize=fontsize, rotation=0)
plt.xticks(x, labels, fontsize=fontsize, rotation=0)

font_prop = font_manager.FontProperties(size=fontsize)
#plt.xlabel('Penalty regularizer $\lambda_{up}$', fontproperties=font_prop)
plt.ylabel('Î”N@10', fontproperties=font_prop)

plt.legend(['Vanilla', 'LinearMix','PDSR'], loc='upper right', prop={'size': fontsize})
plt.grid(True)
plt.tight_layout()
#plt.show()
plt.savefig('out-of-domain.pdf')
```
